```python
fruits_protein = {
    "Banana": 1.1,
    "Apple": 0.3,
    "Orange": 0.9,
    "Mango": 0.8,
    "Guava": 2.6,
    "Avocado": 2.0,
    "Blackberries": 1.4,
    "Pomegranate": 1.7,
    "Papaya": 0.5,
    "Grapes": 0.6
}
```

```python
s2 = pd.Series(fruits_protein,name = 'Proteins')
s2
```

```python
# Coditional selection
s2[s2>1]
```

```python
#Logical operator & | not

s2[(s2 > 1.5) & (s2 <= 2.5)]
```

```python
s2[(s2 > 1.5) | (s2 > 4)]
```

```python
# Modyfing the series

s2['Mango']=10
s2
```
